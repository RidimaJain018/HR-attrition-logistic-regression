```python
import numpy as np
import pandas as pd
```

```python
df1 = pd.read_csv("/kaggle/input/datasets/vjchoudhary7/hr-analytics-case-study/employee_survey_data.csv")
df2 = pd.read_csv("/kaggle/input/datasets/vjchoudhary7/hr-analytics-case-study/general_data.csv")
df3 = pd.read_csv("/kaggle/input/datasets/vjchoudhary7/hr-analytics-case-study/in_time.csv")
df4 = pd.read_csv("/kaggle/input/datasets/vjchoudhary7/hr-analytics-case-study/manager_survey_data.csv")
df5 = pd.read_csv("/kaggle/input/datasets/vjchoudhary7/hr-analytics-case-study/out_time.csv")
```

```python
df3 = df3.rename(columns={'Unnamed: 0': 'EmployeeID'})
df5 = df5.rename(columns={'Unnamed: 0': 'EmployeeID'})
```

```python
df6 = pd.merge(df1, df4, on = 'EmployeeID')
df7 = pd.merge(df6, df2, on = 'EmployeeID')
df8 = pd.merge(df3, df5, on = 'EmployeeID')
df_final = pd.merge(df7, df8, on = 'EmployeeID')
```

```python
df_final.head()
```

```python
df_final.isnull().sum()
```

```python
mode_job = df_final['JobSatisfaction'].mode()[0]
df_final['JobSatisfaction'] = df_final['JobSatisfaction'].fillna(mode_job)
```

```python
mode_env = df_final['EnvironmentSatisfaction'].mode()[0]
df_final['EnvironmentSatisfaction'] = df_final['EnvironmentSatisfaction'].fillna(mode_env)
```

```python
mode_work = df_final['WorkLifeBalance'].mode()[0]
df_final['WorkLifeBalance'] = df_final['WorkLifeBalance'].fillna(mode_work)
```

```python
mode_comp = df_final['NumCompaniesWorked'].mode()[0]
df_final['NumCompaniesWorked'] = df_final['NumCompaniesWorked'].fillna(mode_comp)
```

```python
mode_year = df_final['TotalWorkingYears'].mode()[0]
df_final['TotalWorkingYears'] = df_final['TotalWorkingYears'].fillna(mode_year)
```

```python
def rename_attendance(col):
    if col.endswith('_x'):
        date = col.replace('_x','')
        return f'In_{date}'
    elif col.endswith('_y'):
        date = col.replace('_y','')
        return f'Out_{date}'
    else:
        return col

df_final.columns = [rename_attendance(col) for col in df_final.columns]
```

```python
nan_percentage = df_final.isna().mean()
cols_to_drop = nan_percentage[nan_percentage > 0.30].index
df_final = df_final.drop(columns=cols_to_drop)
```

```python
print(len(cols_to_drop))
print(cols_to_drop)
print(df_final.shape)
```

```python
attendance_cols = [col for col in df_final.columns if col.startswith('In_') or col.startswith('Out_')]
df_final[attendance_cols] = df_final[attendance_cols].applymap(
    lambda x: str(x).split(' ')[-1] if pd.notna(x) else x
)
```

```python
in_cols = [col for col in attendance_cols if col.startswith('In_')]
```

```python
def time_to_minutes(t):
    if pd.isna(t):
        return t
    h, m, s = map(int, t.split(':'))
    return h*60 + m

df_final[attendance_cols] = df_final[attendance_cols].applymap(time_to_minutes)
```

```python
out_cols = [col for col in attendance_cols if col.startswith('Out_')]
work_cols = []

for in_col in in_cols:
    
    date = in_col.replace('In_', '')
    out_col = f'Out_{date}'
    
    if out_col in df_final.columns:
        
        work_col = f'Work_{date}'
        
        df_final[work_col] = df_final[out_col] - df_final[in_col]
        
        work_cols.append(work_col)

df_final['avg_work_hours'] = df_final[work_cols].mean(axis=1)
df_final['avg_work_hours'] = df_final['avg_work_hours'] / 60
```

```python
in_cols = [col for col in attendance_cols if col.startswith('In_')]
all_in_times = df_final[in_cols].stack()
mode_in_time = all_in_times.mode()[0]
```

```python
in_cols = [col for col in attendance_cols if col.startswith('In_')]
df_final['avg_in_time'] = df_final[in_cols].mean(axis=1)
df_final['avg_in_time'] = df_final['avg_in_time'].round().astype(int)
```

```python
df_final[['avg_in_time']].head()
```

```python
df_final['work_hour_std'] = df_final[work_cols].std(axis=1)
df_final['work_hour_std'] = df_final['work_hour_std'] / 60
```

```python
df_final[['work_hour_std']].head()
```

```python
late_matrix = df_final[in_cols] > mode_in_time

df_final['late_ratio'] = late_matrix.sum(axis=1) / df_final[in_cols].notna().sum(axis=1)
```

```python
df_final['attendance_rate'] = df_final[in_cols].notna().sum(axis=1) / len(in_cols)
```

```python
df_final['Attrition'] = df_final.pop('Attrition')
```

```python
cols_to_remove = in_cols + out_cols + work_cols

df_final.drop(columns=cols_to_remove, inplace=True)
```

```python
x = df_final.drop('Attrition', axis=1)
y = df_final['Attrition']

from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 42, stratify=y
)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.impute import SimpleImputer
from sklearn.linear_model import LogisticRegression
```

```python
categorical_cols = [
    'BusinessTravel',
    'Department',
    'EducationField',
    'Gender',
    'JobRole',
    'MaritalStatus'
]

numerical_cols = [col for col in x.columns if col not in categorical_cols]
```

```python
num_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy='median')),
    ('scaler', StandardScaler())
])

cat_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('encoder', OneHotEncoder(drop='first'))
])
```

```python
preprocessor = ColumnTransformer([
    ('num', num_pipeline, numerical_cols),
    ('cat', cat_pipeline, categorical_cols)
])
```

```python
model_pipeline = Pipeline([
    ('preprocess', preprocessor),
    ('model', LogisticRegression(max_iter=1000))
])
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
```

```python
model_pipeline.fit(x_train, y_train)
y_pred = model_pipeline.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

print(accuracy_score(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```
